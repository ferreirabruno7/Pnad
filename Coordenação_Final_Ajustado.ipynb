```python
# --------------------------------------------
# 0. Pacotes
# --------------------------------------------
import pandas as pd
import statsmodels.api as sm
import statsmodels.sandbox.regression.gmm as smgmm
import datetime as dt
import os
import numpy as np # For dummy data generation

# --------------------------------------------
# 1. Carregar dados
# --------------------------------------------
dados_path = '/content/dados.csv'
epu_path   = '/content/Brazil_Policy_Uncertainty_Data.csv'

# --- Optional: Create dummy files if they don't exist for testing ---
# --- Remove or comment out this block if you have your actual data files ---
create_dummy_files = False # Set to True to create dummy files
if create_dummy_files:
    if not os.path.exists(dados_path):
        print(f"Warning: File {dados_path} not found. Creating a dummy file for demonstration.")
        if not os.path.exists('/content'): os.makedirs('/content', exist_ok=True)
        _dates_dados = pd.to_datetime(pd.date_range(start='2000-01-01', end='2023-12-31', freq='M'))
        _num_periods_dados = len(_dates_dados)
        _data_dados = {
            'Date': _dates_dados,
            'ibcbr': np.random.rand(_num_periods_dados) * 10 + 100,
            'ipca12': np.random.rand(_num_periods_dados) * 2 + 4,
            'r_expost': np.random.rand(_num_periods_dados) * 1 + 1,
            'selic_over': np.random.rand(_num_periods_dados) * 3 + 8,
            'primario_pib': np.random.rand(_num_periods_dados) * 1 - 0.5,
            'dlsp_pib': np.random.rand(_num_periods_dados) * 10 + 50,
            'usd_compra': np.random.rand(_num_periods_dados) * 1 + 4
        }
        pd.DataFrame(_data_dados).to_csv(dados_path, index=False)
        print(f"Dummy {dados_path} created.")

    if not os.path.exists(epu_path):
        print(f"Warning: File {epu_path} not found. Creating a dummy file for demonstration.")
        if not os.path.exists('/content'): os.makedirs('/content', exist_ok=True)
        _dates_epu = pd.to_datetime(pd.date_range(start='2000-01-01', end='2023-12-31', freq='M'))
        _year_epu = _dates_epu.year
        _month_epu = _dates_epu.month
        _num_periods_epu = len(_dates_epu)
        _data_epu = {
            'year': _year_epu,
            'month': _month_epu,
            'Brazil_Policy_Index': np.random.rand(_num_periods_epu) * 50 + 100
        }
        pd.DataFrame(_data_epu).to_csv(epu_path, index=False)
        print(f"Dummy {epu_path} created.")
# --- End of optional dummy file creation ---

# Check if data files exist before attempting to read
if not os.path.exists(dados_path):
    raise FileNotFoundError(f"Data file not found: {dados_path}. Please ensure the file exists or enable dummy file creation.")
if not os.path.exists(epu_path):
    raise FileNotFoundError(f"EPU file not found: {epu_path}. Please ensure the file exists or enable dummy file creation.")

dados   = pd.read_csv(dados_path, parse_dates=['Date'])
epu_raw = pd.read_csv(epu_path)

# Construir série mensal de EPU (fim de mês)
epu_raw['Date'] = pd.to_datetime(
    epu_raw['year'].astype(str) + '-' + epu_raw['month'].astype(str).str.zfill(2) + '-01'
) + pd.offsets.MonthEnd(0)
epu = epu_raw[['Date', 'Brazil_Policy_Index']].rename(
    columns={'Brazil_Policy_Index': 'EPU'}
)

# --------------------------------------------
# 2. Hiato do produto (HP filter λ = 129 600)
# --------------------------------------------
dados = dados.dropna(subset=['ibcbr']).copy()
if dados['ibcbr'].empty:
    raise ValueError("DataFrame is empty after dropping NaNs from 'ibcbr'. Check your 'ibcbr' data.")
cycle, trend = sm.tsa.filters.hpfilter(dados['ibcbr'], lamb=129_600)
dados['hiato'] = (dados['ibcbr'] - trend) / trend * 100

# --------------------------------------------
# 3. Merge datasets
# --------------------------------------------
df = dados.merge(epu, on='Date', how='left')

# --------------------------------------------
# 4. Construir variáveis e defasagens (até 4 lags)
# --------------------------------------------
for var, src in [
    ('Y',   'hiato'),
    ('pi',  'ipca12'),
    ('r_real', 'r_expost'),
    ('i_nom',  'selic_over'),
    ('P',   'primario_pib'),
    ('div', 'dlsp_pib'),
    ('usd', 'usd_compra')
]:
    df[var] = df[src]
    for l in range(1, 5):
        df[f'{var}_lag{l}'] = df[var].shift(l)

# Expectativa de inflação (proxy simples π_{t+1})
df['Et_ipca'] = df['pi'].shift(-1)

# Dummy de crises
def crisis_dummy(date):
    return int(
        (dt.datetime(2008, 7, 1) <= date <= dt.datetime(2009, 6, 30)) or
        (dt.datetime(2014, 4, 1) <= date <= dt.datetime(2016, 9, 30)) or
        (dt.datetime(2020, 4, 1) <= date <= dt.datetime(2021, 6, 30))
    )
df['dummy'] = df['Date'].apply(crisis_dummy)

# Diferenças para instrumentos
df['∆r_real'] = df['r_real'] - df['r_real_lag1']
df['∆P']      = df['P']      - df['P_lag1']
df['∆usd']    = df['usd']    - df['usd_lag1']
df['∆pi']     = df['pi']     - df['pi_lag1']
df['∆Y']      = df['Y']      - df['Y_lag1']
df['∆div']    = df['div']    - df['div_lag1']

# --------------------------------------------
# 5. Função auxiliar IV-GMM
# --------------------------------------------
def run_linear_gmm(dep, exog_vars, instr_vars, drop_cols):
    """
    dep        : str (variável dependente)
    exog_vars  : list [str] (regressores “endógenos” + const)
    instr_vars : list [str] (instrumentos + const)
    drop_cols  : list [str] (para eliminar linhas com NaN)
    """
    # Ensure drop_cols contains unique column names
    unique_drop_cols = list(dict.fromkeys(drop_cols))
    local_df = df.dropna(subset=unique_drop_cols).copy()

    if local_df.empty:
        raise ValueError(f"DataFrame became empty after dropping NaNs for columns: {unique_drop_cols}. "
                         "This often happens if lags reduce the sample size drastically or if "
                         "there's extensive missing data in key variables (e.g., 'Et_ipca' due to shift(-1)). "
                         "Check your data and lag structure.")

    endog = local_df[dep]
    exog  = sm.add_constant(local_df[exog_vars])
    instr = sm.add_constant(local_df[instr_vars])

    # Ensure there are enough observations
    if len(endog) <= exog.shape[1]:
         raise ValueError(f"Not enough observations for GMM estimation after NaN removal. "
                          f"Observations: {len(endog)}, Regressors (incl. const): {exog.shape[1]}. "
                          f"Problematic equation might be related to extensive NaNs in: {unique_drop_cols}")

    res   = smgmm.LinearIVGMM(endog, exog, instr).fit()

    # Tabela formatada
    table = pd.DataFrame({
        'coef':   res.params.round(4),
        't-stat': res.tvalues.round(2),
        'p-val':  res.pvalues.round(4)
    })

    # Teste Hansen J
    stat, pval, df_h = res.jtest()
    return table, stat, pval, df_h

# --------------------------------------------
# 6. Especificações das quatro equações
# --------------------------------------------
equations = {}

# --- IS ---
equations['IS'] = dict(
    dep   = 'Y',
    exog  = ['Y_lag1','r_real','P','EPU','dummy'],
    instr = ['Y_lag2','Y_lag3','Y_lag4',
             'r_real_lag1','r_real_lag2','r_real_lag3',
             'P_lag1','P_lag2','P_lag3',
             '∆r_real','∆P','EPU','dummy']
)

# --- Phillips ---
# Note: 'Et_ipca' (pi.shift(-1)) will cause the last observation to be NaN for Et_ipca.
# This NaN will be dropped, reducing sample size by 1 from the end.
equations['Phillips'] = dict(
    dep   = 'pi',
    exog  = ['pi_lag1','Et_ipca','Y_lag1','usd','EPU','dummy'],
    instr = ['pi_lag2','pi_lag3','pi_lag4',
             'Et_ipca', # Using Et_ipca as its own instrument implies it's strictly exogenous.
             'Y_lag2','Y_lag3',
             'usd_lag1','usd_lag2','usd_lag3',
             '∆usd','EPU','dummy']
)

# --- Taylor ---
equations['Taylor'] = dict(
    dep   = 'i_nom',
    exog  = ['pi','Et_ipca','Y','usd','EPU','dummy'],
    instr = ['pi_lag1','pi_lag2','pi_lag3',
             'Et_ipca', # Using Et_ipca as its own instrument implies it's strictly exogenous.
             'Y_lag1','Y_lag2','Y_lag3',
             'usd_lag1','usd_lag2','usd_lag3',
             '∆pi','∆Y','∆usd','EPU','dummy']
)

# --- Fiscal ---
equations['Fiscal'] = dict(
    dep   = 'P',
    exog  = ['P_lag1','Y_lag1','div','EPU','dummy'],
    instr = ['P_lag2','P_lag3','P_lag4',
             'Y_lag2','Y_lag3',
             'div_lag1','div_lag2','div_lag3',
             '∆div','EPU','dummy']
)

# --------------------------------------------
# 7. Estimar, exibir e armazenar resultados
# --------------------------------------------
results_tables = {}
jstats = []

for name, spec in equations.items():
    print(f"Estimating equation: {name}")
    # Define columns to check for NaNs. This includes all regressors, instruments, and the dependent var.
    # 'Et_ipca' is created by pi.shift(-1), so it will have a NaN for the last period.
    # Lags also reduce available sample at the beginning.
    drop_cols_list = spec['exog'] + spec['instr'] + [spec['dep']]
    if 'const' in drop_cols_list: # 'const' is added later by sm.add_constant
        drop_cols_list = [col for col in drop_cols_list if col != 'const']

    try:
        tbl, stat, pval, df_h = run_linear_gmm(
            dep        = spec['dep'],
            exog_vars  = spec['exog'],
            instr_vars = spec['instr'],
            drop_cols  = drop_cols_list
        )
        results_tables[name] = tbl
        jstats.append([name, stat, pval, df_h])

        # Mostrar os resultados (substituindo display_dataframe_to_user)
        print(f"\n--- {name} GMM Results ---")
        print(tbl)
        print(f"Hansen J-test for {name}: stat={stat:.4f}, p-val={pval:.4f}, df={df_h}")

    except ValueError as e:
        print(f"Could not estimate {name}. Error: {e}")
        print(f"  Dependent var: {spec['dep']}")
        print(f"  Exogenous vars: {spec['exog']}")
        print(f"  Instrumental vars: {spec['instr']}")
        # Optionally, print info about NaNs for the problematic equation
        # nan_check_df = df[list(dict.fromkeys(drop_cols_list))].isnull().sum()
        # print(f"NaN counts for columns in {name} before dropna:\n{nan_check_df[nan_check_df > 0]}")
        # temp_df_for_check = df.dropna(subset=list(dict.fromkeys(drop_cols_list)))
        # print(f"Shape of data for {name} after dropna: {temp_df_for_check.shape}")


# Tabela de Hansen J
if jstats: # Only proceed if some estimations were successful
    jstats_df = pd.DataFrame(jstats, columns=['Equation','Hansen J','p-val','df'])
    print("\n--- Overall Over-ID tests (Hansen J) ---")
    print(jstats_df.round(4))
else:
    print("\nNo GMM estimations were successful, so no Hansen J-test summary table.")


# --------------------------------------------
# 8. Exportar LaTeX
# --------------------------------------------
if results_tables: # Only proceed if some tables were generated
    latex_snippets = []
    for key, tbl in results_tables.items():
        caption = f"\\textbf{{{key}}} – Linear IV-GMM (instrumentos estendidos)"
        # Added escape=False to correctly render LaTeX commands like \textbf
        latex_snippets.append(tbl.to_latex(index=True,
                                           caption=caption,
                                           label=key.lower()+'_gmm',
                                           escape=False))

    full_tex = "\n\\clearpage\n".join(latex_snippets)

    # Adjusted output path to be relative (saves in current directory)
    # Original: out_path = '/mnt/data/gmm_results.tex'
    out_path = './gmm_results.tex'

    try:
        with open(out_path, 'w', encoding='utf-8') as texfile:
            texfile.write(full_tex)
        print(f"\nLaTeX salvo em: {os.path.abspath(out_path)}")
    except Exception as e:
        print(f"\nError saving LaTeX file to {out_path}: {e}")
else:
    print("\nNo results tables were generated, so no LaTeX file created.")
```

```python
# --------------------------------------------
# 0. Pacotes
# --------------------------------------------
import pandas as pd
import statsmodels.api as sm
import statsmodels.sandbox.regression.gmm as smgmm
import datetime as dt
import os
import numpy as np # For dummy data generation
import matplotlib.pyplot as plt # <--- Adicionado para gráficos
import matplotlib.dates as mdates # <--- Adicionado para formatar datas

# --------------------------------------------
# 1. Carregar dados
# --------------------------------------------
dados_path = '/content/dados.csv'
epu_path   = '/content/Brazil_Policy_Uncertainty_Data.csv'

# --- Optional: Create dummy files if they don't exist for testing ---
create_dummy_files = False # Set to True to create dummy files
if create_dummy_files:
    if not os.path.exists(dados_path):
        print(f"Warning: File {dados_path} not found. Creating a dummy file for demonstration.")
        if not os.path.exists('/content'): os.makedirs('/content', exist_ok=True)
        _dates_dados = pd.to_datetime(pd.date_range(start='2000-01-01', end='2023-12-31', freq='M'))
        _num_periods_dados = len(_dates_dados)
        _data_dados = {
            'Date': _dates_dados,
            'ibcbr': np.random.rand(_num_periods_dados) * 10 + 100,
            'ipca12': np.random.rand(_num_periods_dados) * 2 + 4,
            'r_expost': np.random.rand(_num_periods_dados) * 1 + 1,
            'selic_over': np.random.rand(_num_periods_dados) * 3 + 8,
            'primario_pib': np.random.rand(_num_periods_dados) * 1 - 0.5,
            'dlsp_pib': np.random.rand(_num_periods_dados) * 10 + 50,
            'usd_compra': np.random.rand(_num_periods_dados) * 1 + 4
        }
        pd.DataFrame(_data_dados).to_csv(dados_path, index=False)
        print(f"Dummy {dados_path} created.")

    if not os.path.exists(epu_path):
        print(f"Warning: File {epu_path} not found. Creating a dummy file for demonstration.")
        if not os.path.exists('/content'): os.makedirs('/content', exist_ok=True)
        _dates_epu = pd.to_datetime(pd.date_range(start='2000-01-01', end='2023-12-31', freq='M'))
        _year_epu = _dates_epu.year
        _month_epu = _dates_epu.month
        _num_periods_epu = len(_dates_epu)
        _data_epu = {
            'year': _year_epu,
            'month': _month_epu,
            'Brazil_Policy_Index': np.random.rand(_num_periods_epu) * 50 + 100
        }
        pd.DataFrame(_data_epu).to_csv(epu_path, index=False)
        print(f"Dummy {epu_path} created.")
# --- End of optional dummy file creation ---

if not os.path.exists(dados_path):
    raise FileNotFoundError(f"Data file not found: {dados_path}. Please ensure the file exists or enable dummy file creation.")
if not os.path.exists(epu_path):
    raise FileNotFoundError(f"EPU file not found: {epu_path}. Please ensure the file exists or enable dummy file creation.")

dados   = pd.read_csv(dados_path, parse_dates=['Date'])
epu_raw = pd.read_csv(epu_path)

epu_raw['Date'] = pd.to_datetime(
    epu_raw['year'].astype(str) + '-' + epu_raw['month'].astype(str).str.zfill(2) + '-01'
) + pd.offsets.MonthEnd(0)
epu = epu_raw[['Date', 'Brazil_Policy_Index']].rename(
    columns={'Brazil_Policy_Index': 'EPU'}
)

# --------------------------------------------
# 2. Hiato do produto (HP filter λ = 129 600)
# --------------------------------------------
dados = dados.dropna(subset=['ibcbr']).copy()
if dados['ibcbr'].empty:
    raise ValueError("DataFrame is empty after dropping NaNs from 'ibcbr'. Check your 'ibcbr' data.")
cycle, trend = sm.tsa.filters.hpfilter(dados['ibcbr'], lamb=129_600)
dados['hiato'] = (dados['ibcbr'] - trend) / trend * 100
dados['ibcbr_trend'] = trend # Guardar a tendência para plotar junto com o ibcbr

# --------------------------------------------
# 3. Merge datasets
# --------------------------------------------
df = dados.merge(epu, on='Date', how='left')

# --------------------------------------------
# 4. Construir variáveis e defasagens (até 4 lags)
# --------------------------------------------
for var, src in [
    ('Y',   'hiato'),
    ('pi',  'ipca12'),
    ('r_real', 'r_expost'),
    ('i_nom',  'selic_over'),
    ('P',   'primario_pib'),
    ('div', 'dlsp_pib'),
    ('usd', 'usd_compra')
]:
    df[var] = df[src]
    for l in range(1, 5):
        df[f'{var}_lag{l}'] = df[var].shift(l)

df['Et_ipca'] = df['pi'].shift(-1)

def crisis_dummy(date):
    return int(
        (dt.datetime(2008, 7, 1) <= date <= dt.datetime(2009, 6, 30)) or
        (dt.datetime(2014, 4, 1) <= date <= dt.datetime(2016, 9, 30)) or
        (dt.datetime(2020, 4, 1) <= date <= dt.datetime(2021, 6, 30))
    )
df['dummy'] = df['Date'].apply(crisis_dummy)

df['∆r_real'] = df['r_real'] - df['r_real_lag1']
df['∆P']      = df['P']      - df['P_lag1']
df['∆usd']    = df['usd']    - df['usd_lag1']
df['∆pi']     = df['pi']     - df['pi_lag1']
df['∆Y']      = df['Y']      - df['Y_lag1']
df['∆div']    = df['div']    - df['div_lag1']

# --------------------------------------------
# 4.5 Visualização dos Dados Temporais       <--- NOVA SEÇÃO
# --------------------------------------------
# Selecionar as colunas para plotar e seus títulos
# Usamos as colunas originais antes de serem renomeadas para Y, pi, etc., para clareza,
# mas também podemos usar as renomeadas se preferir.
# Adicionaremos 'Date' ao df_plot para o eixo x.

df_plot = df.copy() # Usar uma cópia para não alterar df se necessário

cols_to_plot = {
    'ibcbr': 'IBC-Br e Tendência HP',
    'hiato': 'Hiato do Produto (%)',
    'ipca12': 'Inflação Acum. 12m (IPCA %a.a.)',
    'Et_ipca': 'Expectativa de Inflação (Proxy t+1, %a.a.)',
    'r_expost': 'Taxa de Juros Real Ex-Post (%a.a.)',
    'selic_over': 'Taxa Selic Over (%a.a.)',
    'primario_pib': 'Resultado Primário (% PIB)',
    'dlsp_pib': 'Dívida Líquida (% PIB)',
    'usd_compra': 'Taxa de Câmbio (R$/USD Compra)',
    'EPU': 'Índice de Incerteza Política (EPU)',
    'dummy': 'Dummy de Crises (1=Crise)'
}

num_plots = len(cols_to_plot)
num_cols = 2 # Quantidade de colunas de subplots
num_rows = (num_plots + num_cols - 1) // num_cols # Calcula o número de linhas necessário

fig, axes = plt.subplots(num_rows, num_cols, figsize=(15, num_rows * 4))
axes = axes.flatten() # Transforma o array 2D de eixos em 1D para facilitar o loop

# Períodos de crise para sombreamento
crisis_periods_def = [
    (dt.datetime(2008, 7, 1), dt.datetime(2009, 6, 30)),
    (dt.datetime(2014, 4, 1), dt.datetime(2016, 9, 30)),
    (dt.datetime(2020, 4, 1), dt.datetime(2021, 6, 30))
]

for i, (col, title) in enumerate(cols_to_plot.items()):
    ax = axes[i]
    if col == 'ibcbr': # Plot especial para IBC-Br com sua tendência
        ax.plot(df_plot['Date'], df_plot['ibcbr'], label='IBC-Br (Observado)', color='blue')
        if 'ibcbr_trend' in df_plot.columns:
             ax.plot(df_plot['Date'], df_plot['ibcbr_trend'], label='Tendência HP', color='red', linestyle='--')
        ax.legend()
    elif col == 'dummy': # Plot especial para dummy
        ax.plot(df_plot['Date'], df_plot[col], label=title, color='black', linestyle='-')
        ax.fill_between(df_plot['Date'], 0, df_plot[col], alpha=0.3, color='gray', step='post')
        ax.set_yticks([0, 1])
        ax.set_yticklabels(['Não Crise', 'Crise'])
    else:
        ax.plot(df_plot['Date'], df_plot[col], label=title, color='teal')

    ax.set_title(title)
    ax.set_xlabel('Data')
    ax.grid(True, linestyle=':', alpha=0.7)

    # Adicionar sombreamento para períodos de crise em todos os gráficos, exceto o da dummy
    if col != 'dummy':
        for start_date, end_date in crisis_periods_def:
            ax.axvspan(start_date, end_date, color='grey', alpha=0.2, lw=0)

    # Melhorar a formatação das datas no eixo x
    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y-%m'))
    ax.tick_params(axis='x', rotation=45)

# Remover eixos extras se o número de plots não preencher todos os subplots
for j in range(i + 1, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('Visualização das Séries Temporais Macroeconômicas', fontsize=16, y=1.02)
plt.tight_layout(rect=[0, 0, 1, 0.98]) # Ajusta o layout para não sobrepor títulos
plt.show()


# --------------------------------------------
# 5. Função auxiliar IV-GMM
# --------------------------------------------
def run_linear_gmm(dep, exog_vars, instr_vars, drop_cols):
    """
    dep        : str (variável dependente)
    exog_vars  : list [str] (regressores “endógenos” + const)
    instr_vars : list [str] (instrumentos + const)
    drop_cols  : list [str] (para eliminar linhas com NaN)
    """
    unique_drop_cols = list(dict.fromkeys(drop_cols))
    local_df = df.dropna(subset=unique_drop_cols).copy()

    if local_df.empty:
        raise ValueError(f"DataFrame became empty after dropping NaNs for columns: {unique_drop_cols}. "
                         "This often happens if lags reduce the sample size drastically or if "
                         "there's extensive missing data in key variables (e.g., 'Et_ipca' due to shift(-1)). "
                         "Check your data and lag structure.")

    endog = local_df[dep]
    exog  = sm.add_constant(local_df[exog_vars])
    instr = sm.add_constant(local_df[instr_vars])

    if len(endog) <= exog.shape[1]:
         raise ValueError(f"Not enough observations for GMM estimation after NaN removal. "
                          f"Observations: {len(endog)}, Regressors (incl. const): {exog.shape[1]}. "
                          f"Problematic equation might be related to extensive NaNs in: {unique_drop_cols}")

    res   = smgmm.LinearIVGMM(endog, exog, instr).fit()

    table = pd.DataFrame({
        'coef':   res.params.round(4),
        't-stat': res.tvalues.round(2),
        'p-val':  res.pvalues.round(4)
    })

    stat, pval, df_h = res.jtest()
    return table, stat, pval, df_h

# --------------------------------------------
# 6. Especificações das quatro equações
# --------------------------------------------
equations = {}

# --- IS ---
equations['IS'] = dict(
    dep   = 'Y',
    exog  = ['Y_lag1','r_real','P','EPU','dummy'],
    instr = ['Y_lag2','Y_lag3','Y_lag4',
             'r_real_lag1','r_real_lag2','r_real_lag3',
             'P_lag1','P_lag2','P_lag3',
             '∆r_real','∆P','EPU','dummy']
)

# --- Phillips ---
equations['Phillips'] = dict(
    dep   = 'pi',
    exog  = ['pi_lag1','Et_ipca','Y_lag1','usd','EPU','dummy'],
    instr = ['pi_lag2','pi_lag3','pi_lag4',
             'Et_ipca',
             'Y_lag2','Y_lag3',
             'usd_lag1','usd_lag2','usd_lag3',
             '∆usd','EPU','dummy']
)

# --- Taylor ---
equations['Taylor'] = dict(
    dep   = 'i_nom',
    exog  = ['pi','Et_ipca','Y','usd','EPU','dummy'],
    instr = ['pi_lag1','pi_lag2','pi_lag3',
             'Et_ipca',
             'Y_lag1','Y_lag2','Y_lag3',
             'usd_lag1','usd_lag2','usd_lag3',
             '∆pi','∆Y','∆usd','EPU','dummy']
)

# --- Fiscal ---
equations['Fiscal'] = dict(
    dep   = 'P',
    exog  = ['P_lag1','Y_lag1','div','EPU','dummy'],
    instr = ['P_lag2','P_lag3','P_lag4',
             'Y_lag2','Y_lag3',
             'div_lag1','div_lag2','div_lag3',
             '∆div','EPU','dummy']
)

# --------------------------------------------
# 7. Estimar, exibir e armazenar resultados
# --------------------------------------------
results_tables = {}
jstats = []

for name, spec in equations.items():
    print(f"Estimating equation: {name}")
    drop_cols_list = spec['exog'] + spec['instr'] + [spec['dep']]
    if 'const' in drop_cols_list:
        drop_cols_list = [col for col in drop_cols_list if col != 'const']

    try:
        tbl, stat, pval, df_h = run_linear_gmm(
            dep        = spec['dep'],
            exog_vars  = spec['exog'],
            instr_vars = spec['instr'],
            drop_cols  = drop_cols_list
        )
        results_tables[name] = tbl
        jstats.append([name, stat, pval, df_h])

        print(f"\n--- {name} GMM Results ---")
        print(tbl)
        print(f"Hansen J-test for {name}: stat={stat:.4f}, p-val={pval:.4f}, df={df_h}")

    except ValueError as e:
        print(f"Could not estimate {name}. Error: {e}")

if jstats:
    jstats_df = pd.DataFrame(jstats, columns=['Equation','Hansen J','p-val','df'])
    print("\n--- Overall Over-ID tests (Hansen J) ---")
    print(jstats_df.round(4))
else:
    print("\nNo GMM estimations were successful, so no Hansen J-test summary table.")

# --------------------------------------------
# 8. Exportar LaTeX
# --------------------------------------------
if results_tables:
    latex_snippets = []
    for key, tbl in results_tables.items():
        caption = f"\\textbf{{{key}}} – Linear IV-GMM (instrumentos estendidos)"
        latex_snippets.append(tbl.to_latex(index=True,
                                           caption=caption,
                                           label=key.lower()+'_gmm',
                                           escape=False))

    full_tex = "\n\\clearpage\n".join(latex_snippets)
    out_path = './gmm_results.tex'

    try:
        with open(out_path, 'w', encoding='utf-8') as texfile:
            texfile.write(full_tex)
        print(f"\nLaTeX salvo em: {os.path.abspath(out_path)}")
    except Exception as e:
        print(f"\nError saving LaTeX file to {out_path}: {e}")
else:
    print("\nNo results tables were generated, so no LaTeX file created.")
```

```python
# --------------------------------------------
# 0. Pacotes
# --------------------------------------------
import pandas as pd
import numpy as np
import statsmodels.api as sm
import statsmodels.sandbox.regression.gmm as smgmm
import datetime as dt
import os
import matplotlib.pyplot as plt
import matplotlib.dates as mdates

# -------------------------------------------------
# 0.5 Configurações e Placeholders de Dados
# -------------------------------------------------
# Diretório base para dados e saída
base_dir = '/content/'
dados_path = os.path.join(base_dir, 'dados.csv')
consumidor_path = os.path.join(base_dir, 'consumidor.csv')
out_path_tex = os.path.join(base_dir, 'gmm_results_icc.tex')


# --- Optional: Create dummy files if they don't exist for testing ---
create_dummy_files = False # Set to True to create dummy files
if create_dummy_files:
    if not os.path.exists(base_dir):
        os.makedirs(base_dir, exist_ok=True)
        print(f"Diretório {base_dir} criado.")

    # Dummy dados.csv
    if not os.path.exists(dados_path):
        print(f"Warning: File {dados_path} not found. Creating a dummy file.")
        _dates_dados = pd.to_datetime(pd.date_range(start='2000-01-01', end='2024-12-31', freq='M'))
        _num_periods_dados = len(_dates_dados)
        _data_dados = {
            'Date': _dates_dados,
            'ibcbr': np.random.rand(_num_periods_dados) * 10 + 100,
            'ipca12': np.random.rand(_num_periods_dados) * 2 + 4,
            'r_expost': np.random.rand(_num_periods_dados) * 1 + 1,
            'selic_over': np.random.rand(_num_periods_dados) * 3 + 8,
            'primario_pib': np.random.rand(_num_periods_dados) * 1 - 0.5,
            'dlsp_pib': np.random.rand(_num_periods_dados) * 10 + 50,
            'usd_compra': np.random.rand(_num_periods_dados) * 1 + 4
        }
        pd.DataFrame(_data_dados).to_csv(dados_path, index=False)
        print(f"Dummy {dados_path} created.")

    # Dummy consumidor.csv (com o novo formato problemático)
    if not os.path.exists(consumidor_path):
        print(f"Warning: File {consumidor_path} not found. Creating a dummy file with format ['Data', 'iec_t', 'Unnamed: 2'].")
        _dummy_dates_icc = pd.to_datetime(pd.date_range(start='2000-01-01', end='2024-12-31', freq='M'))
        _num_periods_icc = len(_dummy_dates_icc)
        _data_icc = {
            # Formato de data como string para simular o problema
            'Data': [d.strftime('%Y.%m') for d in _dummy_dates_icc], # Ex: 2000.01
            'iec_t': np.random.rand(_num_periods_icc) * 30 + 80,
            'Unnamed: 2': [None] * _num_periods_icc # Coluna extra
        }
        pd.DataFrame(_data_icc).to_csv(consumidor_path, index=False)
        print(f"Dummy {consumidor_path} created.")
# --- End of optional dummy file creation ---

if not os.path.exists(dados_path):
    raise FileNotFoundError(f"Data file not found: {dados_path}. Please ensure the file exists or enable dummy file creation.")
if not os.path.exists(consumidor_path):
    raise FileNotFoundError(f"ICC file not found: {consumidor_path}. Please ensure the file exists or enable dummy file creation.")

# -------------------------------------------------
# 1. Carregar dados macro (dados.csv) e ICC (consumidor.csv)
# -------------------------------------------------
dados = pd.read_csv(dados_path, parse_dates=["Date"])

# Ajuste para carregar consumidor.csv com as colunas ['Data', 'iec_t', 'Unnamed: 2']
icc_raw = pd.read_csv(consumidor_path)
print(f"Colunas originais do consumidor.csv: {icc_raw.columns.tolist()}")

# Identificar e renomear colunas com base no erro e suposições
# Coluna 0: Data/Período (Ex: 'Data')
# Coluna 1: Valor do ICC (Ex: 'iec_t')
# Coluna 2: Ignorar (Ex: 'Unnamed: 2')

if len(icc_raw.columns) >= 2:
    col_periodo_original = icc_raw.columns[0]
    col_icc_original = icc_raw.columns[1]

    icc_temp = pd.DataFrame()
    icc_temp['period_str'] = icc_raw[col_periodo_original].astype(str)
    icc_temp['ICC_value'] = icc_raw[col_icc_original]

    print(f"Usando coluna '{col_periodo_original}' como período e '{col_icc_original}' como ICC do arquivo {consumidor_path}.")
else:
    raise ValueError(f"O arquivo {consumidor_path} não tem pelo menos 2 colunas. Colunas encontradas: {icc_raw.columns.tolist()}")

# Tentar converter a coluna de período para Datetime
try:
    # Tentativa 1: Conversão direta para datetime
    icc_temp['Date'] = pd.to_datetime(icc_temp['period_str'])
    # Se a conversão direta não gerar erro, mas as datas não forem fim de mês, ajustar
    if not icc_temp['Date'].dt.is_month_end.all():
        icc_temp['Date'] = icc_temp['Date'] + pd.offsets.MonthEnd(0)
    print(f"Coluna de período '{col_periodo_original}' convertida diretamente para datetime (fim de mês).")
except Exception: # Usar Exception mais genérico aqui pois o erro pode ser de tipo ou formato
    print(f"Conversão direta de '{col_periodo_original}' para datetime falhou. Tentando formatos AAAA.MM, AAAA-MM, AAAAMM...")
    try:
        # Tenta remover caracteres não numéricos e dividir por '.' ou '-'
        period_cleaned = icc_temp['period_str'].str.replace(r'[^\d\.\-\/]', '', regex=True) # Adicionado /

        # Tenta com '.' como separador
        if period_cleaned.str.contains(r'^\d{4}\.\d{1,2}$', regex=True).all(): # Ex: 2000.01 ou 2000.1
            year_month_split = period_cleaned.str.split(".", expand=True)
            print("Formato de período detectado: AAAA.MM")
        # Tenta com '-' como separador (AAAA-MM)
        elif period_cleaned.str.contains(r'^\d{4}\-\d{1,2}$', regex=True).all(): # Ex: 2000-01 ou 2000-1
            year_month_split = period_cleaned.str.split("-", expand=True)
            print("Formato de período detectado: AAAA-MM")
        # Tenta com '/' como separador (AAAA/MM)
        elif period_cleaned.str.contains(r'^\d{4}\/\d{1,2}$', regex=True).all(): # Ex: 2000/01 ou 2000/1
            year_month_split = period_cleaned.str.split("/", expand=True)
            print("Formato de período detectado: AAAA/MM")
        # Tenta inferir se é AAAAMM
        elif period_cleaned.str.contains(r'^\d{6}$', regex=True).all(): # Ex: 200001
            year_month_split = pd.DataFrame({
                0: period_cleaned.str[:4],
                1: period_cleaned.str[4:]
            })
            print("Formato de período detectado: AAAAMM")
        else: # Se nenhuma das heurísticas funcionar, falha
            raise ValueError("Formato de período (AAAA.MM, AAAA-MM, AAAA/MM, AAAAMM) não reconhecido após limpeza.")

        year  = year_month_split[0].astype(int)
        month = year_month_split[1].astype(int)
        icc_temp["Date"] = pd.to_datetime(
            year.astype(str) + "-" + month.astype(str).str.zfill(2) + "-01"
        ) + pd.offsets.MonthEnd(0)
        print("Coluna de período convertida com sucesso a partir do formato string (fim de mês).")
    except Exception as e_format:
        raise ValueError(f"Erro final ao converter 'period_str' para Date no arquivo ICC. "
                         f"Verifique o formato da coluna '{col_periodo_original}'. "
                         f"Conteúdo exemplo: '{icc_temp['period_str'].iloc[0]}'. Erro original: {e_format}")

icc = icc_temp[["Date", "ICC_value"]].rename(columns={"ICC_value": "ICC"})
if icc['ICC'].isnull().all():
    print(f"Aviso: Todos os valores da coluna ICC são NaN após o processamento de {consumidor_path}.")


# -------------------------------------------------
# 2. Processamento Inicial e Merge
# -------------------------------------------------
df_merged = dados.merge(icc, on="Date", how="left")

start_date = pd.to_datetime("2003-01-01")
end_date_target = pd.to_datetime("2025-04-30")

if not df_merged.empty:
    max_data_date = df_merged['Date'].max()
    if end_date_target > max_data_date:
        print(f"Aviso: A data final da amostra ({end_date_target.strftime('%Y-%m-%d')}) "
              f"é posterior à data mais recente nos dados ({max_data_date.strftime('%Y-%m-%d')}). "
              f"Usando {max_data_date.strftime('%Y-%m-%d')} como data final.")
        end_date = max_data_date
    else:
        end_date = end_date_target
else:
    print("Aviso: DataFrame 'df_merged' está vazio antes da filtragem por data.")
    end_date = end_date_target

df = df_merged[(df_merged['Date'] >= start_date) & (df_merged['Date'] <= end_date)].copy()

if df.empty:
    raise ValueError("DataFrame ficou vazio após a filtragem por data. Verifique o período da amostra e a disponibilidade dos dados.")

df = df.dropna(subset=["ibcbr"]).copy()
if df['ibcbr'].empty or len(df['ibcbr']) < 2:
    raise ValueError("Não há dados suficientes de 'ibcbr' após a filtragem para calcular o hiato do produto. "
                     f"Número de observações de 'ibcbr': {len(df['ibcbr'])}")

cycle, trend = sm.tsa.filters.hpfilter(df['ibcbr'], lamb=129_600)
df["hiato"] = (df["ibcbr"] - trend) / trend * 100
df["ibcbr_trend"] = trend

# -------------------------------------------------
# 3. Construir variáveis e defasagens (até 4 lags)
# -------------------------------------------------
vars_to_lag = {
    "Y": "hiato", "pi": "ipca12", "r_real": "r_expost",
    "i_nom": "selic_over", "P": "primario_pib", "div": "dlsp_pib",
    "usd": "usd_compra", "ICC": "ICC" # ICC já está com o nome correto
}

for var, src in vars_to_lag.items():
    if src in df.columns:
        if var != src:
            df[var] = df[src]
        for l in range(1, 5):
            df[f"{var}_lag{l}"] = df[var].shift(l)
    else:
        print(f"Aviso: Coluna fonte '{src}' para variável '{var}' não encontrada no DataFrame.")

df["Et_ipca"] = df["pi"].shift(-1) if "pi" in df.columns else np.nan

def crisis_dummy(d):
    return int(
        (dt.datetime(2008, 7, 1) <= d <= dt.datetime(2009, 6, 30))
        or (dt.datetime(2014, 4, 1) <= d <= dt.datetime(2016, 9, 30))
        or (dt.datetime(2020, 4, 1) <= d <= dt.datetime(2021, 6, 30))
    )
df["dummy"] = df["Date"].apply(crisis_dummy)

delta_vars_to_create = ['r_real', 'P', 'usd', 'pi', 'Y', 'div', 'ICC']
for var_name in delta_vars_to_create:
    if var_name in df.columns and f"{var_name}_lag1" in df.columns:
        df[f"Δ{var_name}"] = df[var_name] - df[f"{var_name}_lag1"]
    else:
        print(f"Aviso: Não foi possível criar Δ{var_name} pois '{var_name}' ou '{var_name}_lag1' não existem ou têm NaNs.")


# -------------------------------------------------
# 3.5 Visualização dos Dados Temporais
# -------------------------------------------------
df_plot = df.copy()
cols_to_plot = {
    'ibcbr': 'IBC-Br e Tendência HP', 'hiato': 'Hiato do Produto (%) [Y]',
    'ipca12': 'Inflação Acum. 12m (%a.a.) [pi]', 'ICC': 'Índice de Confiança do Consumidor [ICC]',
    'r_expost': 'Taxa de Juros Real Ex-Post (%a.a.) [r_real]', 'selic_over': 'Taxa Selic Over (%a.a.) [i_nom]',
    'primario_pib': 'Resultado Primário (% PIB) [P]', 'dlsp_pib': 'Dívida Líquida (% PIB) [div]',
    'dummy': 'Dummy de Crises'
}

num_plots = len(cols_to_plot)
num_cols_fig = 2
num_rows_fig = (num_plots + num_cols_fig - 1) // num_cols_fig

fig, axes = plt.subplots(num_rows_fig, num_cols_fig, figsize=(15, num_rows_fig * 3.5), sharex=False)
axes = axes.flatten()

crisis_periods_def = [
    (dt.datetime(2008, 7, 1), dt.datetime(2009, 6, 30)),
    (dt.datetime(2014, 4, 1), dt.datetime(2016, 9, 30)),
    (dt.datetime(2020, 4, 1), dt.datetime(2021, 6, 30))
]

plot_idx = 0
for col, title in cols_to_plot.items():
    if plot_idx >= len(axes): break
    ax = axes[plot_idx]
    if col not in df_plot.columns:
        print(f"Aviso de plotagem: Coluna '{col}' não encontrada no DataFrame.")
        ax.set_title(f"{title}\n(Dados não encontrados)", color='red')
        ax.plot([],[])
    elif df_plot[col].isnull().all():
        print(f"Aviso de plotagem: Coluna '{col}' contém apenas NaNs.")
        ax.set_title(f"{title}\n(Apenas NaNs)", color='orange')
        ax.plot([],[])
    else:
        if col == 'ibcbr':
            ax.plot(df_plot['Date'], df_plot['ibcbr'], label='IBC-Br', color='blue')
            if 'ibcbr_trend' in df_plot.columns:
                 ax.plot(df_plot['Date'], df_plot['ibcbr_trend'], label='Tendência HP', color='red', linestyle='--')
            ax.legend()
        elif col == 'dummy':
            ax.plot(df_plot['Date'], df_plot[col], color='black', linestyle='-')
            ax.fill_between(df_plot['Date'], 0, df_plot[col], alpha=0.3, color='gray', step='post')
            ax.set_yticks([0, 1])
        else:
            ax.plot(df_plot['Date'], df_plot[col], color='teal')
        ax.set_title(title)

    ax.grid(True, linestyle=':', alpha=0.7)
    if col != 'dummy' and col in df_plot.columns and not df_plot[col].isnull().all():
        for start_c, end_c in crisis_periods_def:
            ax.axvspan(pd.to_datetime(start_c), pd.to_datetime(end_c), color='grey', alpha=0.2, lw=0)

    ax.xaxis.set_major_formatter(mdates.DateFormatter('%Y'))
    ax.tick_params(axis='x', rotation=30)
    plot_idx += 1


for j in range(plot_idx, len(axes)):
    fig.delaxes(axes[j])

fig.suptitle('Séries Temporais Macroeconômicas e ICC (Amostra Selecionada)', fontsize=16, y=1.03)
plt.tight_layout(rect=[0, 0, 1, 0.99])
plt.show()

# -------------------------------------------------
# 4. Função auxiliar: rodar IV-GMM
# -------------------------------------------------
def gmm_equation(dep_var, exog_vars, instr_vars, data_frame):
    needed_cols = list(dict.fromkeys(exog_vars + instr_vars + [dep_var]))
    missing_in_df = [col for col in needed_cols if col not in data_frame.columns]
    if missing_in_df:
        raise ValueError(f"As seguintes colunas necessárias para GMM não existem no DataFrame: {missing_in_df}")

    panel = data_frame.dropna(subset=needed_cols).copy()

    if panel.empty:
        raise ValueError(f"DataFrame vazio após dropna para colunas: {needed_cols}. Verifique NaNs e lags.")

    Y = panel[dep_var]
    X = sm.add_constant(panel[exog_vars])
    Z = sm.add_constant(panel[instr_vars])

    if len(Y) <= X.shape[1]:
        raise ValueError(f"Observações insuficientes ({len(Y)}) para {X.shape[1]} regressores (incl. constante).")

    if Z.shape[1] < X.shape[1]:
        raise ValueError(f"Número de instrumentos ({Z.shape[1]}) deve ser maior ou igual ao número de regressores ({X.shape[1]}).")


    model = smgmm.LinearIVGMM(Y, X, Z).fit()
    j_stat, j_pval, j_df = model.jtest()

    table = pd.DataFrame(
        {
            "coef": model.params,
            "t-stat": model.tvalues,
            "p-val": model.pvalues,
        },
        index=X.columns,
    )
    return table, j_stat, j_pval, j_df, len(panel)


# -------------------------------------------------
# 5. Especificações das quatro equações (ICC no lugar de EPU)
# -------------------------------------------------
specs = {
    "IS": {"dep": "Y", "ex": ["Y_lag1", "r_real", "P", "ICC", "dummy"]},
    "Phillips": {"dep": "pi", "ex": ["pi_lag1", "Et_ipca", "Y_lag1", "usd", "ICC", "dummy"]},
    "Taylor": {"dep": "i_nom", "ex": ["pi", "Et_ipca", "Y", "usd", "ICC", "dummy"]},
    "Fiscal": {"dep": "P", "ex": ["P_lag1", "Y_lag1", "div", "ICC", "dummy"]},
}

delta_vars_available_for_instr = [col for col in df.columns if col.startswith("Δ")]

for key, sp in specs.items():
    instr_list = []
    for v_exog in sp["ex"]:
        if v_exog.endswith("_lag1") and v_exog[:-5] in vars_to_lag.keys():
            base_var_name = v_exog[:-5]
            for l in range(2, 5):
                instr_list.append(f"{base_var_name}_lag{l}")

    fixed_instruments_candidates = delta_vars_available_for_instr + ["ICC", "dummy"]
    if "Et_ipca" in sp["ex"]:
         fixed_instruments_candidates.append("Et_ipca")

    instr_list.extend(fixed_instruments_candidates)

    final_instruments_for_spec = []
    seen_instruments = set()
    for inst_candidate in instr_list:
        if inst_candidate in df.columns:
            if inst_candidate not in seen_instruments:
                final_instruments_for_spec.append(inst_candidate)
                seen_instruments.add(inst_candidate)
        else:
            print(f"Aviso de instrumento: '{inst_candidate}' para eq '{key}' não encontrado no DF e será omitido.")
    specs[key]["instr"] = final_instruments_for_spec


# -------------------------------------------------
# 6. Rodar as quatro equações
# -------------------------------------------------
results = {}
estimations_summary = []

for eq_name, sp_content in specs.items():
    print(f"\nEstimando equação: {eq_name}")

    all_needed_vars_for_eq = list(dict.fromkeys(sp_content["ex"] + sp_content["instr"] + [sp_content["dep"]]))
    missing_vars_in_df = [v for v in all_needed_vars_for_eq if v not in df.columns]

    if missing_vars_in_df:
        error_msg = f"Variáveis faltando no DataFrame para a equação {eq_name}: {missing_vars_in_df}."
        print(f"Aviso: {error_msg} Pulando esta equação.")
        results[eq_name] = {"table": pd.DataFrame(), "error": error_msg}
        estimations_summary.append([eq_name, 0, np.nan, np.nan, 0, error_msg])
        continue

    try:
        tbl, j_stat, j_pval, j_df, nobs = gmm_equation(
            sp_content["dep"], sp_content["ex"], sp_content["instr"], df
        )
        results[eq_name] = {
            "table": tbl.round(4), "J_stat": round(j_stat, 3),
            "J_pval": round(j_pval, 3), "J_df": j_df, "N_obs": nobs,
        }
        estimations_summary.append([eq_name, nobs, round(j_stat,3), round(j_pval,3), j_df, "OK"])
    except ValueError as e:
        error_msg_val = str(e)
        print(f"Erro ao estimar {eq_name}: {error_msg_val}")
        results[eq_name] = {"table": pd.DataFrame(), "error": error_msg_val}
        estimations_summary.append([eq_name, 0, np.nan, np.nan, 0, error_msg_val])


# -------------------------------------------------
# 7. Exibir resultados (print simples)
# -------------------------------------------------
for eq_name, res_content in results.items():
    print(f"\n=== {eq_name} ===")
    if "error" in res_content and res_content["error"]:
        print(f"  Erro: {res_content['error']}")
    elif not res_content.get("table", pd.DataFrame()).empty:
        print(f"  N = {res_content['N_obs']}")
        print(res_content["table"])
        print(f"  Hansen J: {res_content['J_stat']} (df={res_content['J_df']}) p-value = {res_content['J_pval']}")
    else:
        print("  Nenhuma tabela de resultado gerada ou erro durante a estimação.")

summary_df = pd.DataFrame(estimations_summary, columns=['Equation', 'N_obs', 'Hansen_J', 'p_J', 'df_J', 'Status'])
print("\n\n--- Resumo das Estimações GMM ---")
print(summary_df)

# -------------------------------------------------
# 8. Exportar LaTeX
# -------------------------------------------------
if any(not res.get("table", pd.DataFrame()).empty for res in results.values()):
    latex_snippets = []
    for eq_name, res_content in results.items():
        if not res_content.get("table", pd.DataFrame()).empty:
            caption = (f"\\textbf{{{eq_name}}} – IV-GMM com ICC (N={res_content['N_obs']}) \\\\ \n"
                       f"Hansen J: {res_content['J_stat']} (p-val: {res_content['J_pval']}, df: {res_content['J_df']})")
            latex_snippets.append(res_content["table"].to_latex(index=True,
                                           caption=caption,
                                           label=f"tab:{eq_name.lower()}_icc_gmm",
                                           escape=False,
                                           column_format='lrrr'))

    full_tex = "\n\\clearpage\n".join(latex_snippets)

    try:
        if not os.path.exists(base_dir) and base_dir == '/content/':
             print(f"Aviso: Diretório {base_dir} não existe. Tentando criar...")
             try:
                 os.makedirs(base_dir, exist_ok=True)
                 print(f"Diretório {base_dir} criado com sucesso.")
             except Exception as e_mkdir:
                 print(f"Falha ao criar diretório {base_dir}: {e_mkdir}. O arquivo LaTeX pode não ser salvo.")

        with open(out_path_tex, 'w', encoding='utf-8') as texfile:
            texfile.write(full_tex)
        print(f"\nLaTeX salvo em: {out_path_tex}")
    except Exception as e:
        print(f"\nErro ao salvar arquivo LaTeX em {out_path_tex}: {e}")
else:
    print("\nNenhuma tabela de resultados foi gerada, arquivo LaTeX não criado.")
```
